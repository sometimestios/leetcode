# 题目描述
两个整数之间的汉明距离指的是这两个数字对应二进制位不同的位置的数目。
给出两个整数 x 和 y，计算它们之间的汉明距离。
注意：
0 <= x, y < 2^31.

[461.汉明距离](https://leetcode-cn.com/problems/hamming-distance/solution/yi-ming-ju-chi-by-leetcode/)

## 方法一：我的方法
暴力数数

```python
x=4
y=1
def hamming(x,y):
    x=list(bin(x))[2:]
    x=[int(i) for i in x]
    y=list(bin(y))[2:]
    y=[int(i) for i in y]
    if len(x)>len(y):
        y=[0 for i in range(len(x)-len(y))]+y
    else:
        x=[0 for i in range(len(y)-len(x))]+x
    count=0
    for i,j in zip(x,y):
        if i^j:
            count+=1
    return count
hamming(x,y)
```

## 方法二：count方法
即使转换成二进制数的位数不一样，也可以直接作异或运算；
用str类型自带的count方法返回1的个数

```python
x=4
y=1
def hamming(x,y):
    r=bin(x^y)
    return r.count('1')
hamming(x,y)
```

## 方法三：布赖恩·克尼根算法（神奇）


```python
x=4
y=1
def hamming(x,y):
    r=x^y
    count=0
    while r:
        if r &1:  # 末尾为1时
            count+=1
        r=r>>1
    return count
hamming(x,y)
```
